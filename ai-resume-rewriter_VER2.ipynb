```python
import os
from langchain_groq import ChatGroq

# REPLACE THIS with your actual API key
os.environ["GROQ_API_KEY"] = "YOUR API KEY"

# Initialize the Model
llm = ChatGroq(
    model="llama-3.3-70b-versatile", 
    temperature=0.3
)

# Quick test to make sure it's working
try:
    test_response = llm.invoke("Say 'Hello, Ajay!' if you can hear me.")
    print(test_response.content)
except Exception as e:
    print(f"Error: {e}")
```

```python
input_bullets = [
    "Scheduled appointments, managed emails and calls for senior leadership.",
    "Provided end-to-end event logistics and meeting support.",
    "Familiarity with audit techniques, processes, and tools.",
    "Proficiency in using GST portals for registration, payment, and return filing.",
    "Extracted, cleaned & integrated multi-source datasets (Insurance Records, Customer Feedback); performed data profiling & executed sentiment analysis via Power Query to deliver customer-centric insights.",
    "Designed & deployed dynamic dashboards using Bar, Line, Ribbon, Donut & Matrix visualizations; implemented Role-Based Security (RLS) & scheduled automated data refresh for real-time, secure reporting."
]
```

```python
from langchain_core.prompts import ChatPromptTemplate
from langchain_core.output_parsers import StrOutputParser

# Strategy 1: Role-Playing (The Expert Recruiter)
prompt_role = ChatPromptTemplate.from_template(
    """
    Role: You are an expert Resume Writer with 10 years of experience in HR.
    Task: Rewrite the following resume bullet point to be professional, concise, and impactful.
    Constraint: Start with a strong action verb. Keep it between 12-20 words.
    
    Original Bullet: {input_text}
    
    Rewritten Bullet:
    """
)

# Strategy 2: Direct Instruction (Concise & Technical)
prompt_direct = ChatPromptTemplate.from_template(
    """
    Task: Polish this resume bullet point.
    Requirements:
    1. Start with a power verb (e.g., Engineered, Orchestrated, Analyzed).
    2. Remove all fluff.
    3. Include industry keywords.
    4. Make it sound sophisticated but clear.
    
    Original Bullet: {input_text}
    
    Rewritten Bullet:
    """
)

# Strategy 3: Few-Shot (Giving Examples)
prompt_few_shot = ChatPromptTemplate.from_template(
    """
    Rewrite the input bullet point following these examples of "Bad" vs "Good":
    
    Bad: "Sold stuff to clients."
    Good: "Generated $50k in revenue by executing targeted sales strategies for key clients."
    
    Bad: "Handled code for the app."
    Good: "Optimized application performance by refactoring 1,000+ lines of legacy Python code."
    
    Input Bullet: {input_text}
    
    Rewritten Bullet:
    """
)

# Create Chains
chain_role = prompt_role | llm | StrOutputParser()
chain_direct = prompt_direct | llm | StrOutputParser()
chain_few_shot = prompt_few_shot | llm | StrOutputParser()
```

```python
results = []

print("Processing bullets... (This usually takes 5-10 seconds)")

for bullet in input_bullets:
    try:
        # Generate outputs using the 3 strategies
        out_role = chain_role.invoke({"input_text": bullet})
        out_direct = chain_direct.invoke({"input_text": bullet})
        out_few_shot = chain_few_shot.invoke({"input_text": bullet})
        
        # Store in a dictionary
        results.append({
            "Original Input": bullet,
            "Strategy 1 (Role-Based)": out_role,
            "Strategy 2 (Direct)": out_direct,
            "Strategy 3 (Few-Shot)": out_few_shot
        })
    except Exception as e:
        print(f"Error processing bullet: {bullet[:20]}... | Error: {e}")

print("Processing complete!")
```

```python
import pandas as pd

# Create a DataFrame
df = pd.DataFrame(results)

# Display appropriately in Jupyter (showing full text)
pd.set_option('display.max_colwidth', None)
display(df)

# Save to CSV 
df.to_csv("Mployee_Task_Output.csv", index=False)
```
